```python
import cv2
import numpy as np
import os

# Input and Output Folder Paths
input_folder = "C:/Users/HP/Videos/FINAL_Project/Guava/MINE/Normal/Unhealthy"  # Change to your image folder path
output_folder = "C:/Users/HP/Videos/FINAL_Project/Guava/MINE/Normal/Unhealthy"

# Create output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Function to apply augmentations
def augment_image(image, image_name):
    (h, w) = image.shape[:2]

    # 1. Right Shift
    M_right = np.float32([[1, 0, 50], [0, 1, 0]])  # Shift right by 50 pixels
    right_shifted = cv2.warpAffine(image, M_right, (w, h))
    cv2.imwrite(os.path.join(output_folder, f"{image_name}_right.jpg"), right_shifted)

    # 2. Left Shift
    M_left = np.float32([[1, 0, -50], [0, 1, 0]])  # Shift left by 50 pixels
    left_shifted = cv2.warpAffine(image, M_left, (w, h))
    cv2.imwrite(os.path.join(output_folder, f"{image_name}_left.jpg"), left_shifted)

    # 3. Horizontal Flip
    flipped = cv2.flip(image, 1)
    cv2.imwrite(os.path.join(output_folder, f"{image_name}_flipped.jpg"), flipped)

    # 4. Rotation (30 degrees)
    center = (w // 2, h // 2)
    M_rot = cv2.getRotationMatrix2D(center, 30, 1.0)
    rotated = cv2.warpAffine(image, M_rot, (w, h))
    cv2.imwrite(os.path.join(output_folder, f"{image_name}_rotated.jpg"), rotated)

# Process all images in the folder
for filename in os.listdir(input_folder):
    if filename.endswith((".jpg", ".png", ".jpeg")):
        image_path = os.path.join(input_folder, filename)
        image = cv2.imread(image_path)
        
        if image is not None:
            image_name = os.path.splitext(filename)[0]  # Remove file extension
            augment_image(image, image_name)

print("Image augmentation complete! Check the 'augmented_images' folder.")
```
