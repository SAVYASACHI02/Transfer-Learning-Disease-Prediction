```python
import os
import re
import tensorflow as tf
from tensorflow.keras.applications import VGG19, DenseNet121
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Concatenate, Input
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras import backend as K
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc

# ——— GPU & XLA CONFIG —————————————————————————————————————
# Disable XLA auto-JIT for more predictable performance/debugging
os.environ['TF_XLA_FLAGS'] = '--tf_xla_auto_jit=0'
# Enable mixed precision if supported
from tensorflow.keras import mixed_precision
mixed_precision.set_global_policy('mixed_float16')

gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print(f"Using GPU: {[gpu.name for gpu in gpus]}")
    except RuntimeError as e:
        print(f"GPU setup error: {e}")
else:
    print("No GPU found. Using CPU instead.")


def clear_memory():
    """Clear TensorFlow session and free GPU memory."""
    K.clear_session()
    tf.keras.backend.clear_session()
    import gc; gc.collect()

# ——— CHECKPOINT UTIL ————————————————————————————————————
def get_latest_checkpoint(dir_path):
    files = [f for f in os.listdir(dir_path) if f.startswith("model_checkpoint_amla_opt2_epoch_")]
    if not files:
        return None, 0
    epochs = [int(re.search(r'epoch_(\d+)', f).group(1)) for f in files]
    idx = max(range(len(files)), key=lambda i: epochs[i])
    return os.path.join(dir_path, files[idx]), epochs[idx]

# ——— USER SETTINGS —————————————————————————————————————
selected_optimizer = 'adamw'
train_dir      = './amla_images/train'
validation_dir = './amla_images/val'
test_dir       = './amla_images/test'
checkpoint_dir = './checkpoints_amla_opt2'
os.makedirs(checkpoint_dir, exist_ok=True)
checkpoint_path = os.path.join(checkpoint_dir, 'model_checkpoint_amla_opt2_epoch_{epoch:02d}.keras')

# ——— DATA GENERATORS ————————————————————————————————————
batch_size = 2
datagen = ImageDataGenerator(rescale=1./255)
train_gen = datagen.flow_from_directory(
    train_dir, target_size=(224,224), batch_size=batch_size, class_mode='binary')
val_gen   = datagen.flow_from_directory(
    validation_dir, target_size=(224,224), batch_size=batch_size, class_mode='binary')
test_gen  = datagen.flow_from_directory(
    test_dir, target_size=(224,224), batch_size=batch_size,
    class_mode='binary', shuffle=False)

# ——— BUILD MODEL —————————————————————————————————————
# Input and backbones
inp = Input(shape=(224,224,3))
vgg = VGG19(weights='imagenet', include_top=False, input_tensor=inp)
dn  = DenseNet121(weights='imagenet', include_top=False, input_tensor=inp)
# Freeze all layers initially
def freeze_all(model):
    for layer in model.layers:
        layer.trainable = False
freeze_all(vgg)
freeze_all(dn)
# Combine features
x1 = GlobalAveragePooling2D()(vgg.output)
x2 = GlobalAveragePooling2D()(dn.output)
x  = Concatenate()([x1, x2])
x  = Dense(1024, activation='relu')(x)
out= Dense(1, activation='sigmoid', dtype='float32')(x)
model = Model(inputs=inp, outputs=out)

# ——— LOAD FROM CHECKPOINT IF AVAILABLE ———————————————————————
latest_ckpt, last_epoch = get_latest_checkpoint(checkpoint_dir)
if latest_ckpt and os.path.exists(latest_ckpt):
    model = load_model(latest_ckpt)
    print(f"Resuming from epoch {last_epoch}")
else:
    last_epoch = 0
    print("Starting fresh training")

# ——— OPTIMIZER FACTORY ————————————————————————————————————
def get_opt(name, lr):
    if name == 'adamw':
        return tf.keras.optimizers.AdamW(learning_rate=lr, weight_decay=1e-5)
    elif name == 'nadam':
        return tf.keras.optimizers.Nadam(lr)
    else:
        return tf.keras.optimizers.Adam(lr)

# ——— CHECKPOINT CALLBACK ———————————————————————————————————
ckpt_cb = ModelCheckpoint(
    filepath=checkpoint_path,
    save_weights_only=False,
    verbose=1
)

# ——— TRAINING PHASES ————————————————————————————————————
phases = [
    (1e-4, 10, []),        # Phase 0: freeze all
    (1e-5, 5, ['last5']),  # Phase 1: unfreeze last 5 layers
    (1e-6, 5, ['all']),    # Phase 2: unfreeze all
]
histories = []

def unfreeze_layers(model, mode):
    if mode == 'all':
        for layer in model.layers:
            layer.trainable = True
    elif mode == 'last5':
        for layer in model.layers[-5:]:
            layer.trainable = True

for idx, (lr, epochs, flags) in enumerate(phases):
    # Skip if already trained
    if last_epoch >= sum(p[1] for p in phases[:idx+1]):
        continue

    # Unfreeze as requested
    if 'all' in flags:
        unfreeze_layers(model, 'all')
    elif 'last5' in flags:
        unfreeze_layers(model, 'last5')

    # Compile
    model.compile(
        optimizer=get_opt(selected_optimizer, lr),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    # Fit
    history = model.fit(
        train_gen,
        validation_data=val_gen,
        epochs=last_epoch + epochs,
        initial_epoch=last_epoch,
        callbacks=[ckpt_cb],
        steps_per_epoch=train_gen.samples // batch_size,
        validation_steps=val_gen.samples // batch_size
    )
    histories.append(history)
    last_epoch += epochs

    # After each phase except the last, clear memory and reload
    if idx < len(phases) - 1:
        latest_ckpt, _ = get_latest_checkpoint(checkpoint_dir)
        clear_memory()
        model = load_model(latest_ckpt)
        print(f"Cleared session and reloaded model from {latest_ckpt}")

# ——— PLOT HISTORY ————————————————————————————————————
plt.figure()
for i, h in enumerate(histories):
    plt.plot(h.history['accuracy'], label=f'train_acc_phase{i}')
    plt.plot(h.history['val_accuracy'], label=f'val_acc_phase{i}')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()
plt.show()

plt.figure()
for i, h in enumerate(histories):
    plt.plot(h.history['loss'], label=f'train_loss_phase{i}')
    plt.plot(h.history['val_loss'], label=f'val_loss_phase{i}')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()

# ——— FINAL EVALUATION —————————————————————————————————
val_loss, val_acc = model.evaluate(val_gen)
test_loss, test_acc = model.evaluate(test_gen)
print(f"Validation → loss={val_loss:.4f}, acc={val_acc:.4f}")
print(f"Test       → loss={test_loss:.4f}, acc={test_acc:.4f}")

test_gen.reset()
probs = model.predict(test_gen, steps=test_gen.samples // batch_size + 1).ravel()
preds = (probs > 0.5).astype(int)
labels = test_gen.classes
print(classification_report(labels, preds))
cm = confusion_matrix(labels, preds)
print("Confusion Matrix:\n", cm)
fpr, tpr, _ = roc_curve(labels, probs)
roc_auc = auc(fpr, tpr)
print(f"ROC AUC = {roc_auc:.4f}")

plt.figure()
plt.plot(fpr, tpr)
plt.plot([0,1], [0,1], linestyle='--')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve')
plt.show()

# Clear memory one last time\clear_memory()
clear_memory()
```

```python
import os
import re
import tensorflow as tf
from tensorflow.keras.applications import VGG19, DenseNet121
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Concatenate, Input
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras import backend as K
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc

# GPU/MEMORY
os.environ['TF_XLA_FLAGS']='--tf_xla_enable_xla_devices'
tf.keras.mixed_precision.set_global_policy('mixed_float16')

def clear_memory():
    K.clear_session(); tf.keras.backend.clear_session(); import gc; gc.collect()

# Settings
selected_optimizer='adam'
train_dir='./amla_images/train'
val_dir='./amla_images/val'
test_dir='./amla_images/test'
ckpt_dir='./checkpoints_amla_opt'
os.makedirs(ckpt_dir,exist_ok=True)
ckpt_path=os.path.join(ckpt_dir,'model_checkpoint_amla_opt_epoch_{epoch:02d}.keras')

# Data prep
batch_size=4
dg=ImageDataGenerator(rescale=1./255)
train_gen=dg.flow_from_directory(train_dir,(224,224),batch_size,'binary')
val_gen  =dg.flow_from_directory(val_dir,(224,224),batch_size,'binary')
test_gen =dg.flow_from_directory(test_dir,(224,224),batch_size,'binary',shuffle=False)

# Model build
inp=Input((224,224,3)); v=VGG19('imagenet',False,inp); dn= DenseNet121('imagenet',False,inp)
for l in v.layers: l.trainable=False
for l in dn.layers: l.trainable=False
x=Concatenate()([GlobalAveragePooling2D()(v.output),GlobalAveragePooling2D()(dn.output)])
x=Dense(1024,activation='relu')(x)
out=Dense(1,activation='sigmoid',dtype='float32')(x)
model=Model(inp,out)

# Checkpoint helper
def latest_ckpt(dir,prefix):
    fs=[f for f in os.listdir(dir) if prefix in f]
    if not fs: return None,0
    es=[int(re.search(r'epoch_(\d+)',f).group(1)) for f in fs]
    idx=es.index(max(es)); return os.path.join(dir,fs[idx]),es[idx]

lckpt,last_epoch=latest_ckpt(ckpt_dir,'amla_opt_epoch')
if lckpt: model=load_model(lckpt); print(f"Loaded epoch {last_epoch}")
else: last_epoch=0; print("Starting fresh")

# Optimizer
def get_opt(name,lr):
    if name=='adamw': return tf.keras.optimizers.AdamW(lr,1e-5)
    if name=='nadam': return tf.keras.optimizers.Nadam(lr)
    return tf.keras.optimizers.Adam(lr)

# Callback
cb=ModelCheckpoint(ckpt_path,verbose=1)

# Phases
phases=[(1e-4,10,[]),(1e-5,5,['last5']),(1e-6,5,['all'])]
hists=[]
for lr,ep,un in phases:
    if last_epoch>=sum(p[1] for p in phases[:phases.index((lr,ep,un))+1]): continue
    if 'last5' in un:
        for l in v.layers[-5:]: l.trainable=True
        for l in dn.layers[-5:]: l.trainable=True
    if 'all' in un:
        for l in v.layers: l.trainable=True
        for l in dn.layers: l.trainable=True
    model.compile(get_opt(selected_optimizer,lr),'binary_crossentropy',['accuracy'])
    h=model.fit(train_gen,validation_data=val_gen,epochs=last_epoch+ep,initial_epoch=last_epoch,
                steps_per_epoch=train_gen.samples//batch_size,validation_steps=val_gen.samples//batch_size,
                callbacks=[cb])
    hists.append(h)
    last_epoch+=ep

# Plot history
plt.figure();
for h in hists:
    plt.plot(h.history['accuracy']); plt.plot(h.history['val_accuracy'])
plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.title('Acc')
plt.show()

plt.figure();
for h in hists:
    plt.plot(h.history['loss']); plt.plot(h.history['val_loss'])
plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.title('Loss')
plt.show()

# Evaluate & metrics
e_val, a_val = model.evaluate(val_gen)
e_test,a_test = model.evaluate(test_gen)
print(f"Val loss={e_val:.4f}, acc={a_val:.4f}")
print(f"Test loss={e_test:.4f}, acc={a_test:.4f}")

test_gen.reset();
probs=model.predict(test_gen,steps=test_gen.samples//batch_size+1).ravel()
preds=(probs>0.5).astype(int)
labels=test_gen.classes
print(classification_report(labels,preds))
print("Confusion Matrix:\n",confusion_matrix(labels,preds))
fpr,tpr,_=roc_curve(labels,probs)
print(f"ROC AUC={auc(fpr,tpr):.4f}")
plt.figure(); plt.plot(fpr,tpr); plt.plot([0,1],[0,1],linestyle='--')
plt.xlabel('FPR'); plt.ylabel('TPR'); plt.title('ROC')
plt.show()

clear_memory()
```
